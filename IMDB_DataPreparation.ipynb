```python
## Sonuçları dataframe olarak alabilmek için index'i sıfırladık 
## Movie Id için 1'den fazla değer girilen filmleri bulduk, "count" alarak kaç değer girilmiş onu gördük.
movie_update = movie.groupby("title").filter(lambda x: len(x)>1).groupby("title").agg({"movieId":"count"}).reset_index();movie_update
```

```python
## Bir movie'ye ait min ve max değerlerini görüp onları daha sonrasında dublicate edeceğiz
movie_update = movie.groupby("title").filter(lambda x: len(x)>1).groupby("title").agg({"movieId":["min","max"]}).reset_index();movie_update
```

```python
for pm in range(len(movie_update)):
    old_movie_id, new_movie_id = movie_update.loc[pm, [("movieId", "max"), ("movieId", "min")]]

    # Rating veri çerçevesinde movieId güncelleme
    rating.loc[rating["movieId"] == old_movie_id, "movieId"] = new_movie_id

    # Movie veri çerçevesinde genres güncelleme ve duplicate movieId'leri datasetten droplama
    old_genres = movie.loc[movie["movieId"] == new_movie_id, "genres"].values
    if len(old_genres) > 0:
        old_genres = old_genres[0]
    else:
        old_genres = ""
    new_genres = movie.loc[movie["movieId"] == old_movie_id, "genres"].values
    if len(new_genres) > 0:
        new_genres = new_genres[0]
    else:
        new_genres = ""
    movie.loc[movie["movieId"] == new_movie_id, "genres"] += '|' + new_genres
    movie = movie.drop(movie[movie["movieId"] == old_movie_id].index)
```
